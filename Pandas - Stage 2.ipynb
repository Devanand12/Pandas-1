Add a new column 'Total Minutes' 

```python
df['Total Minutes'] = 0
df.dtypes
```

```python
df['Total Minutes'] = df['Day Mins'] + df['Eve Mins'] + df['Night Mins']
df.head(2)
```

Rename the column name to another name

```python
df.rename(columns={'Total Minutes' : 'Total Mins'},inplace=True)
```

```python
df.drop(columns=['Total Mins'], inplace=True)
```
