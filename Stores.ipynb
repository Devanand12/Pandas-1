Drop the outliers by percentile (5% sales volume)

```python
#df.columns
max = df.Item_Outlet_Sales.quantile(.95)
min = df.Item_Outlet_Sales.quantile(.05)
print(max)
print(min)
# value > max or value < min -> deleted
# value < max and value > min -> store
df_data = df[(df.Item_Outlet_Sales < max) & (df.Item_Outlet_Sales > min)]
#df_data
indices = df[(df.Item_Outlet_Sales > max) | (df.Item_Outlet_Sales < min)].index
df.drop(indices)
```

Bin these: 1980,1990,2000,2011 -> "Eighties", "Nineties", "Millennium"

```python
pd.cut(df['Outlet_Establishment_Year'],[1980,1990,2000,2011],labels=["Eighties", "Nineties", "Millennium"]).value_counts()
```
