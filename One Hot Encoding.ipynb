Identify the toss decision

```python
df_matches.toss_decision.value_counts()
df_matches.result.value_counts()
```

Create a new column to find toss decision

```python
df_matches["toss_bat"] = df_matches.toss_decision.map({"field" : 0, "bat" : 1})
```

Another approach - One Hot Encoding (get_dummies)

```python
df_toss = pd.get_dummies(df_matches.toss_decision,prefix='toss',drop_first='true')
pd.concat([df_matches,df_toss],axis=1).head(2)
```
